```python
#Importing requried libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading the dataset
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
#showing the dataset of matches
match.head()
```

```python
#counting the matches
match.shape
```

```python
##showing the dataset of delivery
delivery.head()
```

```python
#show total score grouping by match_id and inning
total_score_df = delivery .groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
#showing Total Runs only inning 1
total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
#show the dataset
total_score_df
```

```python
#merge total_score_df into match file
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
#show the dataset
match_df
```

```python
#showing the teams in matches
match_df['team1'].unique()
```

```python
#store the current playing teams in list team
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
#replace Delhi daredevils into Delhi capitals 
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

#Replace Deccan charges into sunrisers Hyderabad
match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
#Store the teams in team1 and team2 from list team
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df['team1'].unique()
```

```python
match_df.shape
```

```python
#Remove those matces in which DLS applied 
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
#Extract required attribute from matches
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
match_df
```

```python
match_df['winner'].unique()
```

```python
match_df['winner'] = match_df['winner'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['winner'] = match_df['winner'].str.replace('Deccan Chargers','Delhi Capitals')
match_df['winner'] = match_df['winner'].str.replace('Kings XI Punjab','Punjab Kings')
```

```python
match_df['winner'].unique()
```

```python
#merged both dataset using match_id 
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df
```

```python
#counting the runs on every balls
delivery_df['current_score']=delivery_df.groupby(['match_id'])['total_runs_y'].cumsum()
```

```python
delivery_df
```

```python
#Runs left ater every balls deliver
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
#Balls left ater every balls deliver
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
#Player dismissed after the wicked down 
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum()
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
delivery_df.tail()
```

```python
# Current runrate = runs/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
#Required runrate
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

```python
#Result declare in form of 0 & 1
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df['batting_team'] = final_df['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
final_df['bowling_team'] = final_df['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')

final_df['batting_team'] = final_df['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
final_df['bowling_team'] = final_df['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

final_df['batting_team'] = final_df['batting_team'].str.replace('Kings XI Punjab','Punjab Kings')
final_df['bowling_team'] = final_df['bowling_team'].str.replace('Kings XI Punjab','Punjab Kings')
```

```python
final_df['batting_team'].unique()
```

```python
final_df.sample()
```

## EDA ( Exploratory Data Analysis)

```python
# Total Matches Played by Each Team
# We are using matplotlib and seaborn library for data visualization with different graphs and charts

team_matches = final_df['batting_team'].value_counts() + final_df['bowling_team'].value_counts()
plt.figure(figsize = (12,6))
sns.barplot(x = team_matches.index, y = team_matches.values, palette = 'Dark2')
plt.xticks(rotation = 90)
plt.xlabel("Teams")
plt.ylabel("Total Matches Played")
plt.title("Total Matches Played by Each Team")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Total runs scored by each team
team_total_runs = final_df.groupby('batting_team')['total_runs_x'].sum().sort_values()
plt.figure(figsize=(12,6))
sns.barplot(x = team_total_runs.index, y = team_total_runs.values, palette = 'Set1')
plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("Total Runs Scored")
plt.title("Total Runs Scored by Each Team")
plt.show()
```

```python
# Matches played in different cities
city_matches = final_df['city'].value_counts()
plt.figure(figsize = (12,6))
sns.barplot(x = city_matches.index, y=city_matches.values, palette = 'Spectral')
plt.xticks(rotation = 90)
plt.xlabel("Cities")
plt.ylabel("Matches Hosted")
plt.title("Number of Matches Played in Different Cities")
plt.show()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestClassifier())
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)[8]
```
